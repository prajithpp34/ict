```python
# importing pandas & matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Read the dataset to python environment.
iris = pd.read_excel('/content/iris.xls')
iris
```

```python
# 2. Display the columns in the dataset.
iris.columns
```

```python
# shape of object type of given dataframe
iris.info()
```

```python
# summary statistics for numerical columns in DataFrame
iris.describe()
```

```python
# 3. Calculate the mean of each column of the dataset.
print('Mean value for SL = ',iris['SL'].mean())
print('Mean value for SW = ',iris['SW'].mean())
print('Mean value for PL = ',iris['PL'].mean())
print('Mean value for PW = ',iris['PW'].mean())
```

```python
# 4. Check for the null values present in the dataset.
iris.isna().sum()
```

```python
# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.
# 5.1 Distribution - Sepal Length vs Sepal Width
plt.plot(iris.SL,label = 'SL') # line plot for SL
plt.plot(iris.SW,label = 'SW') # line plot for SW
plt.title('Distribution - Sepal Length vs Sepal Width') # Adding title for graph
plt.legend() # for labelling
plt.show() # To display
```

```python
# 5.2 Box Plot for SW, SL, PW, PL
plt.boxplot([iris['SL'],iris['SW'],iris['PL'],iris['PW']], labels = ['SL', 'SW', 'PL', 'PW']) # box plot for SL, SW, PL, PW
plt.title('Box Plot') # Adding title for graph
plt.show() # To display
```

```python
# 5.3 Scatter Plot SW vs SL
plt.scatter(iris['SL'],iris['SW']) # scatter plot for SL vs SW
plt.title('Scatter Plot') # Adding title for graph
plt.xlabel('Sepal Length') # Labelling x-axis
plt.ylabel('Sepal Width') # Labelling y-axis
plt.show() # To display
```

```python
# 5.4 Histogram for SL
plt.hist(iris['SL'],edgecolor='black',color='green') # histogram for SL, color changee
plt.title('Histogram') # Adding title for graph
plt.xlabel('Sepal Length') # Labelling x-axis
plt.ylabel('Frequency') # Labelling y-axis
plt.show() # To display
```
